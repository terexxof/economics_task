- $a = 0,95$ - изменение товарной массы, поставляемой конкурентами на рынок, доли единицы

- $Д = 1,12$ - коэффициент изменения дохода потребителей товара, доли единицы

## Решение
А. Расчет при затратном ценообразовании

1. Определим годовую производственную программу завода по формуле
$$
q = K_{м} \cdot M
$$

```python
q = K_m * M

print(f'Годовая производственная программа q = {int(q/1000)} тыс. шт/год')
```

2. Определим ожидаемую цену одного кирпича по формуле
$$
z = \frac{\alpha_{min} + \alpha_{max}}{2} \cdot Z_{p}
$$

```python
z = (alpha_min + alpha_max)/2 * Z_p

print(f'Ожидаемая цена одного кирпича z = {z} руб./шт')
```

3. Определим годовую выручку от продажи кирпича по формуле

$$
B = q \cdot z
$$

```python
B = q * z

print(f'Годовая выручка от продажи одного кирпича B = {B} руб./год')
```

4. Определим годовые переменные затраты предприятия по формуле
$$
V_{пер} = V \cdot q
$$

```python
V_per = V * q

print(f'Годовые переменные затраты Vпер = {int(V_per)} руб./год')
```

5. Определим условно-постоянные затраты на единицу продукции по формуле
$$
c = \frac{C}{q}
$$

```python
c = C / q

print(f'Условно-постоянные затраты на единицу продукции c = {round(c, 5)} руб./шт')
```

6. Определим себестоимость производства единицы продукции по формуле
$$
s = V + c
$$

```python
s = V + c

print(f'Себестоимость производства единицы продукции s = {round(s,5)} руб./шт')
```

7. Определим себестоимость годовой товарной продукции по формуле
$$
S = s \cdot q
$$

```python
S = s * q

print(f'Себестоимость годовой товарной продукции S = {round(S)} руб./год')
```

8. Определим величину годовой балансовой прибыли предприятия по формуле
$$
П_{б} = q \cdot (z - s)
$$

```python
P_b = q * (z - s)

print(f'Годовая балансовая прибыль Пб = {round(P_b)} руб./год')
```

9. Определим рентабельность изготовления кирпича по формуле
$$
Р_{и} = \frac{z - s}{s}
$$

```python
R_i = (z - s) / s

print(f'Рентабельность изготовления кирпича Ри = {round(R_i * 100, 2)} %')
```

10. Определим годовую производственную программу самоокупаемости деятельности предприятия по формуле
$$
q_{s} = \frac{C}{Z - V}
$$

```python
q_s = C / (z - V)

print(f'Годовая производственная программа самоокупаемости q_s = {round(q_s)} шт./год')
```

11. Определим характеристику и степень надежности будущего бизнеса. Найдем отношение производственной мощности предприятия $M$ к производственной программе самоокупаемости $q_{s}$

```python
relation_M_q_s = M / q_s

print(f'Отношение производственной мощности предприятия к производственной программе самоокупаемости составляет {round(relation_M_q_s, 2)}')
```

*Промежуточный вывод*
Полученная цифра свидетельствует, что бизнес опосредованно учитывает неопределенность будущей рыночной ситуации и будет достаточно надежным, так как его уровень риска ниже среднего, а поправочная норма эффективности капитальных затрат $E_{р}$, учитывающая риск вложений, составляет ???

12. Определим величину годового совокупного налога предприятия (при отсутствии льгот по налогообложению) по формуле
$$
Н = Н_{пос} + Н_{пер} = C \cdot \lambda + П_{б} \cdot \beta
$$

```python
N = C * lambda_tax + P_b * beta_tax

print(f'Годовой совокупный налог предприятия составляет Н = {round(N,2)} руб./год')
```

13. Определим годовую производственную программу самоокупаемости с учетом налогообложения по формуле
$$
q_{C} = \frac{C(1 + \lambda - \beta)}{(Z - V)(1 - \beta)}
$$

```python
q_C = (C + (1 + lambda_tax - beta_tax)) / ((z - V) * (1-beta_tax))

print(f'Годовая производственная программа самоокупаемости с учетом налогообложения q_C = {round(q_C)} шт./год')
```

*Промежуточный вывод*
Полученный результат свидетельствует о том, что с учетом налогообложения производственная программа самоокупаемости значительно возросла (с 3 145 478 до 4 368 722 шт./год), т.е. увеличилась в X раз. Это существенно сокращает величину чистой прибыли, повышает риск вложений в данный бизнес.

14. Определим долю выручки, остающуюся в распоряжении предприятия после выплаты налогов, по формуле
$$
О_{п} = 1 - \frac{q[\beta \cdot (Z - V) + V] + C(1 + \lambda - \beta)}{q \cdot z}
$$

```python
O_p = 1 - (q * (beta_tax * (z - V) + V) + C * (1 + lambda_tax - beta_tax)) / (q * z)

print(f'В распоряжении предприятия после выплаты налогов останется {round(O_p,3)} или {round(O_p * 100, 1)}%')
```
